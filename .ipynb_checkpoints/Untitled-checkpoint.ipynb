```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('day.csv')
df1=df.drop(['dteday','instant'], axis=1)
df1.head()
```

#### From your analysis of the categorical variables from the dataset, what could you infer about their effect on the dependent variable?

```python
# Analysing Season variable with count
seasonDf = df[['season','cnt']].groupby('season', as_index=False).mean()
seasonDf.head()
seasonDf['season']= seasonDf['season'].map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
seasonDf.head()
```

```python
# Barplot for seasons
sns.barplot(data=seasonDf, x='season',y='cnt')
plt.show()
```

The bar plot of the season vs count shows that more cycles that are taken during fall and the least amount of cycles being taken in spring.

```python
# Analysing weather variables with count
weatheritDf = df[['weathersit','cnt']].groupby('weathersit', as_index=False).mean()
weatheritDf.head()
weatheritDf['weathersit']= weatheritDf['weathersit'].map({1:'Clear', 2:'Mist', 3:'Snow', 4:'Rain'})
weatheritDf.head()
```

```python
# Barplot for weather
sns.barplot(data=weatheritDf, x='weathersit',y='cnt')
plt.show()
```

There is a very clear relationship between the Weather and the number of bikes that are rented. Clear days a lot of bikes are rented vs a big dropoff on bikes that are rented when it snows.

```python
# Analysing month variables with count
monthDf = df[['mnth','cnt']].groupby('mnth', as_index=False).mean()
monthDf.head()
monthDf['mnth']= monthDf['mnth'].map({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',7:'Jul',8:'Aug', 9:'Sep',10:'Oct', 11:'Nov', 12:'Dec'})
monthDf.head()
```

```python
# Barplot for Month
sns.barplot(data=monthDf, x='mnth',y='cnt')
plt.show()
```

This graph correlates to the findings from the seasons column. We see that the months corresponding to Summer and Autumn have the highest number of cyclists.

```python
# Analysing weekday variables with count
weekdayDf = df[['weekday','cnt']].groupby('weekday', as_index=False).mean()
weekdayDf.head()
weekdayDf['weekday']= weekdayDf['weekday'].map({1:'Mon', 2:'Tue', 3:'Wed', 4:'Thu', 5:'Fri', 6:'Sat',7:'Sun'})
weekdayDf.head()
```

```python
# Barplot for weekday
sns.barplot(data=weekdayDf, x='weekday',y='cnt')
plt.show()
```

Weekday does not seem to affect the ridership of bikes

```python
# Analysing month variables with count
holidayDf = df[['holiday','cnt']].groupby('holiday', as_index=False).mean()
holidayDf.head()
holidayDf['holiday']= holidayDf['holiday'].map({0:'Not Holiday', 1:'Holiday'})
holidayDf.head()
```

```python
# Barplot for holiday
sns.barplot(data=holidayDf, x='holiday',y='cnt')
plt.show()
```

On an average more cycles are taken on a working day than on a holiday.

### Training Model using Statsmodel to get more insights into the the model

```python
import statsmodels.api as sm 

# Add constant to calculate the intercept
X_train_lm = sm.add_constant(X_train)
```

```python
sm_model = sm.OLS(y_train,X_train_lm).fit()
sm_model.summary()
```

```python
y__train_pred_sm = sm_model.predict(sm.add_constant(X_train))
```

```python
fig = plt.figure()
sns.histplot((y_train - y__train_pred_sm), bins = 20, kde=True)
fig.suptitle('Error Terms', fontsize = 20)  
plt.xlabel('Errors', fontsize = 18)  
plt.show()
```

Errors are follwing normal distribution and are centered around 0

```python
###qqplot for another verification of above stated conclusion
sm.qqplot((y_train - y__train_pred_sm), fit=True, line='45')
plt.show()
```

```python
# Predicting using the model
y_pred_sm = sm_model.predict(sm.add_constant(X_test))
```

```python
print('r2_score of train dataset', r2_score(y_train, y__train_pred_sm))
print('r2_score of test dataset', r2_score(y_test, y_pred_sm))
```

r2 score for both training data and test data is almost equal. This shows that the model has fit quite well and is not overfitted to the training data.

```python
params_df = pd.DataFrame(sm_model.params)
params_df.index.names = ['Independent Variable']
params_df.rename(columns = {0:'Coefficient value'},inplace = True)
params_df.sort_values(by='Coefficient value')
```

Based on the above coefficient values we can conclude that
1. Year - Positive
2. Snowy weather - Negative
3. Winter Season - Positive
are the biggest contributors.

This shows that the business is growing customers as the years pass and customers are more likely to rent a bike when it is cooler but it should not be snowing.
